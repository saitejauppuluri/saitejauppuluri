```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head()
```

# **Version 1: Detailed Approach**

```python
# Select numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Show the first 10 entries of numeric columns before replacing missing values
print("Original Numeric Data:")
print(df[numeric_columns].head(10))

# Replace missing values with the mean for each numeric column
df[numeric_columns] = df[numeric_columns].apply(lambda x: x.fillna(x.mean()))

# Show the first 10 entries of numeric columns after replacing missing values
print("\nAfter Replacing Missing Values (Version 1):")
print(df[numeric_columns].head(10))
```

# **Version 2: Simplified Approach**

```python
# Reload the dataset to start fresh
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')

# Replace missing values with the mean (automatic for all numeric columns)
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Show the first 10 entries of numeric columns after replacing missing values
print("\nAfter Replacing Missing Values (Version 2):")
print(df[numeric_columns].head(10))
```

# **When to Use Each Version**
**Version 1 (Detailed Approach):**

*   Use when you want more control over specific columns.
*   Good for complex datasets where different columns need different handling.
*   Helpful if you plan to do more than just filling missing values.

**Version 2 (Simplified Approach):**

*   Use for quick data cleaning when all numeric columns need the same treatment.
*   Best for small to medium datasets where speed is important.
*   Great if all numeric columns can be handled the same way.
